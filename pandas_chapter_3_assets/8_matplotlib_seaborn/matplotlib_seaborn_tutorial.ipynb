```python
import pandas as pd

df_sales = pd.read_excel("linechart.xlsx")
df_sales.head()
```

## Line Chart

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"], df_sales["Fridge"], color="blue", label="Fridge")
plt.plot(df_sales["Quarter"], df_sales["Dishwasher"], color="orange", label="Dishwasher")
plt.plot(df_sales["Quarter"], df_sales["Washing Machine"], color="green", label="Washing Machine")

plt.title("Product Sales")
plt.xlabel("Quarter")
plt.ylabel("Million $")

plt.legend()

plt.show()
```

## Pie Chart

```python
total_sales = df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sales
```

```python
plt.pie(total_sales, labels=total_sales.index, autopct='%1.2f%%', startangle=140, explode=(0.1,0,0), shadow=True)
plt.show()
```

## Bar Chart

```python
df_sales.plot(kind="bar", x="Quarter")

plt.xticks(rotation=45) #rotate the x labels
plt.show()
```

To use Quarters as X axis, we can set them as an index on a dataframe and then use plot function

```python
df_sales_2 = df_sales.set_index("Quarter") #sets a default index so you don't need to keep passing it in
df_sales_2
```

```python
df_sales_2.plot(kind="bar")
plt.xticks(rotation=45)
plt.show()
```

```python
df_sales_2.plot(kind="barh")
plt.show()
```

## Histogram

```python
df_score = pd.read_excel("histograms.xlsx")
df_score.head(3)
```

```python
plt.hist(df_score["Exam_Score"], density=True)
plt.show()
```

```python
import seaborn as sns

sns.histplot(df_score["Exam_Score"], kde=True) #kde is kernal density estimator

plt.xlabel("Score")
plt.ylabel("Count")
plt.title("Exam Scores")
plt.show()
```
